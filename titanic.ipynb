```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

titanic_df = sns.load_dataset('titanic')
```

```python
titanic_df
```

```python
titanic_df.info()
```

```python
#non 값을 채워주기
titanic_df['age'].fillna(titanic_df['age'].mean(), inplace = True)
titanic_df['embarked'].fillna(titanic_df['embarked'].describe()['top'], inplace = True) 
titanic_df['embark_town'].fillna(titanic_df['embark_town'].describe()['top'], inplace = True) 
titanic_df['deck'].fillna(titanic_df['deck'].describe()['top'], inplace = True) 
```

```python
titanic_df.info()
```

```python
titanic_df
```

```python
titanic_df.describe()
```

```python
#문자형 데이터를 숫자형을 변환
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
titanic_df['sex'] = le.fit(titanic_df['sex']).transform(titanic_df['sex'])
titanic_df['class'] = le.fit(titanic_df['class']).transform(titanic_df['class'])
titanic_df['adult_male'] = le.fit(titanic_df['adult_male']).transform(titanic_df['adult_male'])
titanic_df['alone'] = le.fit(titanic_df['alone']).transform(titanic_df['alone'])
titanic_df
```

```python
#불필요한 데이터 삭제
drop_cols = ["embarked", 'who', 'deck', 'embark_town', 'alive']
titanic_df = titanic_df.drop(drop_cols, axis =1)
```

```python
titanic_df
```

```python
#Train and test and split
X = titanic_df.iloc[:,1:]
Y = titanic_df['survived']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
#KNN예측모델
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```

```python
#DecisionTree예측모델
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```

```python
#RandomForest예측모델
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```

```python
#GaussianNB예측모델
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```

```python
#SVC예측모델
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```

```python
#LogisticRegression예측모델
from sklearn.linear_model._logistic import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```

```python
#AdaBoostClassifier예측모델
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

print('예측 정확도 %.2f' % accuracy_score(Y_test, y_pred))
```
